```python
import pandas as pd

url = "https://raw.githubusercontent.com/GrandmaCan/ML/main/Resgression/Salary_Data.csv"

data = pd.read_csv(url)
data
# y = w*x + b
x = data["YearsExperience"]
y = data["Salary"]
```

```python
w = 20
b = 20
#n = len(x)
#w_gradient = (2*x*(w*x+b -y)) .sum() /n #cost function對w作微分的結果，再把其作加總，除以總資料數 w_gradient = (2*x*(w*x+b -y)).mean()
w_gradient = (2*x*(w*x+b -y)).mean()
#b_gradient = (2*(w*x+b -y)) .sum() /n #cost function對b作微分的結果
b_gradient = (2*(w*x+b -y)) .mean()
w_gradient
```

```python
def compute_gradient(x ,y ,w ,b):
   w_gradient = (x*(w*x+b -y)).mean()
   b_gradient = ((w*x+b -y)) .mean()
   return w_gradient ,b_gradient
```

```python
compute_gradient(x ,y ,20 ,10)
```

```python
def compute_cost(x, y ,w ,b):
   y_pred = w*x +b
   cost = (y-y_pred)**2 #cost 為(真實值-預測值)**2
   cost = cost.sum() / len(x)

   return cost
```

```python
w = 0
b = 0
learning_rate = 0.001
def gradient_descent(x ,y ,w_init ,b_init ,learning_rate ,cost_function ,gradient_function ,run_iter ,p_iter=1000):
  w = w_init
  b = b_init

  c_hist = []
  w_hist = []
  b_hist = []

  for i in range(run_iter):
   w_gradient ,b_gradient = gradient_function(x ,y ,w ,b)

   w = w - w_gradient*learning_rate #持續更新W值
   b = b - b_gradient*learning_rate #持續更新b值
   cost = cost_function(x, y ,w ,b)

   w_hist.append(w)
   b_hist.append(b)
   c_hist.append(cost)

   if i%p_iter == 0:
     print(f"Ieration {i:5} : Cost {cost: .4e} ,w: {w: .4e} ,b: {b: .4e} ,w_gradient: {w_gradient: .4e} ,b_gradient: {b_gradient: .4e}") #{i:3}-排列有3個空位 {a:.5f}-小數點後面有5個數字
  return w ,b ,w_hist ,b_hist ,c_hist
```

```python
w_init = 0
b_init = 0
learning_rate = 1.0e-3
run_iter = 20000

w_final ,b_final ,w_hist ,b_hist ,c_hist = gradient_descent(x ,y ,w_init ,b_init ,learning_rate ,compute_cost ,compute_gradient ,run_iter)
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.plot(np.arange(0 ,60) ,c_hist[ :60]) #x值用np.arange y值用cost 把cost畫成圖來看
plt.title("iteration vs cost")
plt.xlabel("iteration")
plt.ylabel("cost")

plt.show()
```

```python
ws = np.arange(-100 ,101)
bs = np.arange(-100 ,101)
costs = np.zeros([201 ,201])
i = 0
for w in ws:
  j = 0
  for b in bs:
    cost = compute_cost(x, y ,w ,b)
    costs[i,j] = cost
    j = j+1
  i = i+1
```
